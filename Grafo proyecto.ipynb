```python
import networkx as nx 
import os
import matplotlib.pyplot as plt

ruta_eventos = os.path.join("eventos.csv")
archivo_eventos = open(ruta_eventos, "r")

cont = 0
eventos = []
for cada_evento in archivo_eventos.readlines():
    data = cada_evento.strip("\n").split(";")
    eventos.append(data)
    cont += 1
    if cont == 81:
        break

    
archivo_eventos.close()
#print(eventos)

#Diccionario eventos 

#Goal: Crear grafo eventos primer día con coordenadas

#CArgar coordenadas nodos con arcos correspondientes

coordenadas_nodos = {}

ruta_nodos = os.path.join("nodos.csv")
archivo_nodos = open(ruta_nodos, "r")


archivo_nodos.readline()
for linea in archivo_nodos.readlines():
    data = linea.strip("\n").split(";")
    coordenadas_nodos[data[0]] = (float(data[1]), float(data[2])) #diccionario: {id_nodo:(coordenada_X, coordenada_Y)}
  
archivo_nodos.close()

                               ##########           ARCOS          ############

arcos = os.path.join("arcos.csv")
archivo_arcos = open(arcos, "r")

cont = 0
eventos_e = []
for cada_evento in archivo_arcos.readlines():     #### CREACION DE ARCOS  ####
    data_evento = cada_evento.strip("\n").split(";")
    eventos_e.append([data_evento[0],data_evento[1]])     
    cont += 1
eventos_e.pop(0)

#print(len(eventos_e))

archivo_nodos.close()

N = nx.Graph()

contador = 0

for i in coordenadas_nodos:
    N.add_node(i, pos=(coordenadas_nodos[i][0], coordenadas_nodos[i][1]))
    N.add_edge(eventos_e[contador][0],eventos_e[contador][1])                ##### SE METEN LOS ARCOS AL GRAFO  #####
    #print(contador)
    if contador==12222:
        break
  #  try:
    while eventos_e[contador][0]==eventos_e[contador+1][0]:
        if contador==12224:
            break
        else:
            contador+=1
            N.add_edge(eventos_e[contador][0],eventos_e[contador][1])
    else:
        contador+=1

pos = nx.get_node_attributes(N, 'pos')
nx.draw(N, pos, node_size=5) #, node_color = "blue")

#CArgar coordenadas eventos
coordenadas = {}

archivo_eventos = open(ruta_eventos, "r")

cont1 = 0
archivo_eventos.readline()
for linea in archivo_eventos.readlines():
    data = linea.strip("\n").split(";")
    coordenadas[cont1] = (float(data[0]), float(data[1])) #diccionario: {id_evento:(coordenada_X, coordenada_Y)}
    cont1 += 1
    if cont1 == 81:
        break

archivo_eventos.close()

#print(coordenadas)

                                #####          Cargar coordenadas Bases        #####

coordenadas_bases = {}

ruta_bases = os.path.join("bases.csv")
archivo_bases = open(ruta_bases, "r")

cont1 = 0
archivo_bases.readline()
for linea in archivo_bases.readlines():
    data = linea.strip("\n").split(";")
    coordenadas_bases[cont1] = (float(data[0]), float(data[1])) #diccionario: {id_evento:(coordenada_X, coordenada_Y)}
    cont1 += 1
  

archivo_bases.close()

B = nx.Graph()

for i in coordenadas_bases:
    B.add_node(i, pos=(coordenadas_bases[i][0], coordenadas_bases[i][1]))

pos = nx.get_node_attributes(B, 'pos')
nx.draw(B, pos, node_size=30, node_color = "r")



# Grafo de Eventos
G = nx.Graph()

### agregamos los nodos

for i in coordenadas:
    G.add_node(i, pos=(coordenadas[i][0], coordenadas[i][1]))

pos = nx.get_node_attributes(G, 'pos')
nx.draw(G, pos, node_size=20, node_color = "y")


                                        #####    CREACION DE CENTROS  #####
    
    
coordenadas_centros = {}

centros = os.path.join("centros.csv")
archivo_centros = open(centros, "r")

cont1 = 0
archivo_centros.readline()
for linea in archivo_centros.readlines():
    data = linea.strip("\n").split(";")
    coordenadas_centros[cont1] = (float(data[0]), float(data[1])) #diccionario: {id_evento:(coordenada_X, coordenada_Y)}
    cont1 += 1
  

archivo_centros.close()

C = nx.Graph()

for i in coordenadas_centros:
    C.add_node(i, pos=(coordenadas_centros[i][0], coordenadas_centros[i][1]))

pos = nx.get_node_attributes(C, 'pos')
nx.draw(C, pos, node_size=30, node_color = "black")

## nodos: azules
## bases: rojas
## centros: negros
## eventos: verdes (son los de un dia nomas)

plt.show()
```
